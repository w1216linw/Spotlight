# Seaborn
    Seaborn is a high-level library, and was built on top of matplotlib. It provide people with simple code to visualize complex 
    statistical plots and it come with nice visual features, exquisite graphics and colors.
         "If Matplotlib "tries to make easy things easy and hard things possible", seaborn tries to make a well-defined set of 
         hard things easy too." â€“ Michael Waskom (Creator of Seaborn)

# Import seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

#iris is a dataset come with the seabornS
df_iris = sns.load_dataset('iris')
df_iris.head()
```

# Graphs

    Since seaborn was built on top of matplotlib, let us see the difference in the graph made with seaborn and matplotlib. 

```python
# The scatter plot in seaborn
# Set sepal_length as x and petal_length as y
sns.scatterplot(x = 'petal_length' , y = 'petal_width' , data = df_iris)
plt.show()
```

```python
#The scatter plot in matplotlib
# Set sepal_length as x and petal_length as y
plt.scatter(x = 'petal_length' , y = 'petal_width' , data = df_iris)
plt.show()
```

    There is many point stick together and you can clearly see every point in the seaborn, but in the matplotlib there form a line.
    And with same one line code, seaborn has the name for x-axis and y-axis lable.

```python
#The scatter plot with color in seaborn
# Set sepal_length as x, petal_length as y, and color with species.

sns.scatterplot(x = 'petal_length' , y = 'petal_width' , data = df_iris , hue = 'species')
plt.show()
```

```python
#The scatter plot with color in matplotlib.
# Declare a color for every species.
# Set sepal_length as x, petal_length as y, color with species.

color = {'setosa':'blue' , 'versicolor':'orange' , 'virginica':'green'}
plt.scatter(x = 'petal_length' , y = 'petal_width' , data = df_iris, c = df_iris['species'].map(color))
plt.show()
```

    Base on the two plot, seaborn only need one line code but has more information than the matplotlib. 

```python
#The boxplot by seaborn
sns.boxplot(y='petal_length', data = df_iris)
plt.show()
```

```python
#The boxplot by matplotlib
plt.boxplot(x = 'petal_length' , data = df_iris)
plt.show()
```

```python
#The boxplot by matplotlib
sns.boxplot(y='petal_length',x = 'species', data = df_iris)
plt.show()
```

    Seaborn color his boxplot, and can easily form more than one chart in the same plot.

# More graphs in seaborn

```python
#"tips" is one of dataset that come with seaborn
df_tips = sns.load_dataset('tips')
df_tips.head()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df_tips, col = 'day' ,hue = 'time')
plt.show()
```

    Use lmplot() in seaborn can be form a scatter plot with linear regression model on it.

```python
sns.displot(x = 'total_bill', col = 'time', data = df_tips , kde=True)
plt.show()
```

        displot() in seaborn can be used in statistical analyze to gain the distribution of variables in the dataset.

```python
#Calculated the percentage of tips.
df_tips["tip_percentage"] = df_tips["tip"] / df_tips["total_bill"]
df_tips.head()
```

```python
#x axis is tip_percentage, and the y is count.And the color different by lunch time and dinner time.
sns.histplot(x = 'tip_percentage' , data = df_tips, hue = 'time')
plt.show()
```

    Here is the example of finding difference of tip_percentage on the lunch and dinner.

#   Interesting graph from Seaborn official website

```python
#Conditional kernel density estimate
#Code from seaborn official website
sns.set_theme(style="whitegrid")

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot the distribution of clarity ratings, conditional on carat
sns.displot(
    data=diamonds,
    x="carat", hue="cut",
    kind="kde", height=6,
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75"
)
plt.show()
```

    plot shows better quality of a diamond is, the lighter color on the plot.

```python
#Annotated heatmaps
#Code from seaborn official website
sns.set_theme()

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
plt.show()
```

    Plot shows more passengers, the light color in plot.
